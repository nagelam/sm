```python
# Основные библиотеки
import requests
from bs4 import BeautifulSoup
import csv
import re

# Анализ данных и визуализация
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Машинное обучение
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, recall_score, 
                           confusion_matrix, classification_report, 
                           roc_auc_score)

# Объяснение модели
import shap

# Автоматизированный EDA
import sweetviz as sv

# Оптимизация гиперпараметров
import optuna
from catboost import CatBoostClassifier
```

```python
def extract_table_from_html(html):
    soup=BeautifulSoup(html, 'html.parser')
    columns=[]
    table_data=[]
    table_title=soup.find('h2',string='Water Potability Table')
    if not table_title:
        return None,[]
    table=table_title.find_next('table')
    if not table:
        return None,[]
    headers=table.find('tr')
    if headers:
        for th in headers.find_all('th'):
            columns.append(th.get_text(strip=True))
    for row in table.find_all('tr')[1:]:
        data=[]
        for td in row.find_all('td'):
            data.append(td.get_text(strip=True))
        table_data.append(data)
    return columns,table_data


    

def adding_table_from_each_website_to_csv(links):
    page_num=1
    all_data=[]
    all_cols=[]
    for link in links:
        try:
            response=requests.get(link)
            response.encoding = 'utf-8'
            col_names, data=extract_table_from_html(response.text)
            if(page_num==1 and col_names):
                all_cols=col_names
            all_data.extend(data)
            if(page_num%100)==0:
                print(f"{page_num}/{len(links)} page")
            page_num+=1
        except:
            print(f"error in {page_num}")
            page_num+=1
    with open('temp.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(all_cols)
        writer.writerows(all_data)
    

            

def parse_links(url):
    index_page=requests.get(url+"index.html")
    soup=BeautifulSoup(index_page.text, 'html.parser')
    links=[]
    for a in soup.find_all('a',href=True):
        href=a['href']
        links.append(href)
    print("find",len(links),"links")
    adding_table_from_each_website_to_csv(links)


def main(url):
    parse_links(url)

    


url="https://ratcatcher.ru/media/summer_prac/parcing/16/"
main(url)
```

```python
df=df.drop(columns=['gameId', 'redWins'])
```

```python
from sklearn.preprocessing import StandardScaler
from catboost import CatBoostClassifier
```

```python
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df.drop(columns=['blueWins'])), columns=df.drop(columns=['blueWins']).columns)
df_scaled['blueWins']=df['blueWins'].values
```

```python
df_scaled
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(
    df_scaled.drop(columns=['blueWins']), df_scaled['blueWins'], 
    test_size=0.2, 
    random_state=42,
    stratify=df_scaled['blueWins']
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val,
    test_size=0.25,
    random_state=42,
    stratify=y_train_val
)

def objective(trial):
    params = {
        'iterations': trial.suggest_int('iterations', 500, 1500), 
        'learning_rate': trial.suggest_float('learning_rate', 0.005, 0.2, log=True), 
        'depth': trial.suggest_int('depth', 4, 8),  
        'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10), 
        'random_strength': trial.suggest_float('random_strength', 0.1, 1), 
        'bootstrap_type': 'Bayesian', 
        'grow_policy': 'SymmetricTree',
        'eval_metric': 'AUC',
        'verbose': False
    }
    
    model = CatBoostClassifier(**params)
    
    try:
        model.fit(
            X_train, y_train,
            eval_set=(X_val, y_val),
            early_stopping_rounds=100,
            verbose=0
        )
        
        y_pred = model.predict_proba(X_val)[:, 1]
        y_pred = np.clip(y_pred, 1e-15, 1-1e-15)
        
        return roc_auc_score(y_val, y_pred)
    
    except Exception as e:
        print(f"Trial failed with params: {params}")
        print(f"Error: {str(e)}")
        return 0.5 

sampler = optuna.samplers.TPESampler(
    n_startup_trials=10,
    consider_prior=True,
    multivariate=True
)
study = optuna.create_study(
    direction='maximize',
    sampler=sampler
)

study.optimize(objective, n_trials=50, timeout=3600)
```

```python


# 1. Получаем лучшие параметры из Optuna
best_params = study.best_params
print("Лучшие гиперпараметры:")
for key, value in best_params.items():
    print(f"{key}: {value}")

# 2. Обучаем финальную модель на ОБЪЕДИНЕННЫХ train + val данных
final_model = CatBoostClassifier(
    **best_params,
    eval_metric='AUC',
    verbose=100  # Показываем процесс обучения
)

# Объединяем тренировочные и валидационные данные
X_train_val = pd.concat([X_train, X_val])
y_train_val = pd.concat([y_train, y_val])

final_model.fit(
    X_train_val, 
    y_train_val,
    early_stopping_rounds=100
)

# 3. Предсказания на ТЕСТОВОМ наборе (X_test)
y_test_pred = final_model.predict(X_test)
y_test_proba = final_model.predict_proba(X_test)[:, 1]

# 4. Вычисляем метрики на тесте
test_accuracy = accuracy_score(y_test, y_test_pred)
test_recall = recall_score(y_test, y_test_pred)
test_auc = roc_auc_score(y_test, y_test_proba)

print("\n=== Результаты на ТЕСТОВОМ наборе ===")
print(f"Accuracy: {test_accuracy:.4f}")
print(f"Recall: {test_recall:.4f}")
print(f"AUC-ROC: {test_auc:.4f}")

# 5. Confusion Matrix для теста
cm_test = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Test Set Confusion Matrix')
plt.savefig("2_2.png")
plt.show()

# 6. Полный отчет по метрикам для теста
print("\nTest Set Classification Report:")
print(classification_report(y_test, y_test_pred))
```

```python


# Инициализация SHAP
explainer = shap.TreeExplainer(final_model)
shap_values = explainer.shap_values(X_test)

# 1. Summary Plot (Глобальная важность признаков)
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, plot_type="bar", show=False)
plt.title("SHAP Feature Importance (Bar Plot)")
plt.tight_layout()
plt.savefig('2_3_2.png')

plt.show()

# 2. Beeswarm Plot (Распределение влияния признаков)
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_test, show=False)
plt.title("SHAP Value Distribution (Beeswarm Plot)")
plt.tight_layout()
plt.savefig('2_3_1.png')
plt.show()

```
